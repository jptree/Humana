# Preliminary NaN Handling

If number of NaNs in column < 5000, remove rows, else elimnate column.

```python
def NaNHandling(df, NaNCutoff):
    naColumns = df.isnull().sum()[df.isnull().sum()>0]
    for index, value in naColumns.iteritems():
        if value > NaNCutoff:
            df = df.drop(index, axis=1)
        else:
            df = df[~df[index].isna()]
    return df

NaNCutoff = 5000
dfClean = NaNHandling(df, NaNCutoff)
```

```python
rowsLost = df.shape[0] - dfClean.shape[0]
print ('rows lost: {}'.format(rowsLost))
columnsLost = df.shape[1] - dfClean.shape[1]
print ('columns lost: {}'.format(columnsLost))
```

Transform Categorical Variables

```python
from sklearn import preprocessing

def prepare_inputs(X):
    oe = preprocessing.OrdinalEncoder()
    oe.fit(X)
    X__enc = oe.transform(X)
    return pd.DataFrame(X__enc)
```

```python
# Scale Data (0,1)
# Create X DF
X = dfClean.drop(['transportation_issues'], axis=1)
predictors = X.columns

# prepare input data (change categorical variables to numbers)
X__enc = prepare_inputs(X)

# Convert all columns to numeric
#Xcols = X__enc.columns
#df[Xcols] = df[Xcols].apply(pd.to_numeric, errors='coerce')

# Scale Data (0,1)
X = pd.DataFrame(preprocessing.scale(X__enc, axis=1), columns=X__enc.columns)

#Create Y DF
y = dfClean['transportation_issues']

# Split Data into Testing and Training
X_train, X_test = train_test_split(X, test_size=0.25, random_state=50, shuffle=True)
y_train, y_test = train_test_split(y, test_size=0.25, random_state=50, shuffle=True)
```

```python

predictors
```

```python
from sklearn.linear_model import LassoCV
from sklearn.linear_model import Lasso
from sklearn.model_selection import KFold

n_folds = 10
k_fold = KFold(n_folds)

lasso_cv = LassoCV(alphas=None, cv=k_fold, max_iter=100000)
lasso_cv.fit(X_train, y_train)
print(lasso_cv.alpha_)

lasso = Lasso(alpha=lasso_cv.alpha_, random_state=50, max_iter=100000)
lasso.fit(X_train, y_train)
pred_test_lasso = lasso.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
print('MAE:', mean_absolute_error(y_test, pred_test_lasso))
# print(lasso.coef_)
for i,x in enumerate(list(predictors)):
    print(x, lasso.coef_[i])
```
